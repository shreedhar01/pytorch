## Tensor data type

**Note:** Tensor datatype is one of the 3 error you will encounter with Pytorch and deep learning
1. Tensor not right datatype
2. Tensor not right shape
3. Tensor not on the right device

```python
float_32 = torch.tensor(
    [3.1, 6, 9],
    dtype=None,
    device=None,  #which device is your tensor on
    requires_grad=False   #to track gradient or not with this tensor
  )

float_32, float_32.dtype, float_32.device
```

```python
float_32.shape, float_32.ndim, float_32.requires_grad
```

```python
#changing from 32 to 16

float_16 = float_32.type(torch.float16)
float_16,float_16.dtype, float_16.device
```

```python
a = float_32 * float_16
a, a.dtype, a.device
```

```python
b = float_16 * float_32
b, b.dtype, b.device
```

```python
a = torch.rand(size=( 3, 4))
b = torch.rand(size=( 4, 3))
```

```python
a @ b
```

```python
%%time
torch.matmul(a, b)
```

```python
%%time
result = torch.zeros(size=(a.shape[0], b.shape[1]))
for i in range(len(a)):
  for j in range(len(b[0])):
    for k in range(len(b)):
      result[i][j] += a[i][k] * b[k][j]

result
```
